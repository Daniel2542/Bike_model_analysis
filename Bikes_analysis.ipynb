Number of companies per country

```python
# companies_per_country = df.groupby("Company")["Country of Origin"].sum().sort_values(ascending=False)
# companies_per_country.head(30)

companies_per_country = df.groupby("Company")["Country of Origin"].value_counts()
companies_per_country.head(30)
```

Number of models per country

```python
company_country = df.groupby("Country of Origin").agg(
    total_bikes = ("Model", "size"),
    unique_companies = ("Company", "nunique")
).reset_index()

company_country["bikes_per_company"] = company_country["total_bikes"] / company_country["unique_companies"]

company_country.sort_values("bikes_per_company", ascending=False)
```
